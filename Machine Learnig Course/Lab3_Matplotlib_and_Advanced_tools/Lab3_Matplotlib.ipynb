### 1.1.2 Line plot

````python
# use the first approach for now, since we simply need one plot

plt.figure()

# plt.plot(x-coordinates, y-coordinates)
# the coordinates can be a list or a numpy array or ...

# straight line between two points 
plt.plot([1, 2], [1, 2]) 

# lines between multiple points 
plt.plot([1, 2, 3], [1, 1.5, 1]) 
x = [1, 2, 3, 4]
y = [2, 2.5, 2, 3]
plt.plot(x, y)

plt.show()
# in this simple case, you don't even need plt.figure() and plt.show(), the .plot() method can work independently
````

### 1.1.3 Property of plot

````python
# for the same figure, how to specify labels, add title, ...?

plt.figure()

# in the .plot(), you change the property of the target line
# color, line style, point style (marker), thickness, ...
# refer to https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib-pyplot-plot
plt.plot(
    [1, 2], [1, 2], 
    label = 'line_1',
    color = 'black',
    linewidth = 8
) 
plt.plot(
    [1, 2, 3], [1, 1.5, 1], 
    label = 'line_2',
    linestyle = '--',
    marker = 'o'
) 
x = [1, 2, 3, 4]
y = [2, 2.5, 2, 3]
plt.plot(x, y)

# add title
plt.title('Title of example plot')

# name x and y axis
plt.xlabel('x-label')
plt.ylabel('y-label')

# add (activate) legend
# the label of each line should be specified in the .plot()
# only two legends are shown, since the label keyword is not specified in the third line
plt.legend()

# add (activate) grid
plt.grid()

plt.show()
````

### 1.1.4 Create multiple plots

````python
# now, we need to refer to the second approach to create multiple plots.
# each plot has its own axes and you need to specify the axes to make plot.
# the way you manipulate the plot is similar to the previous approach with slight differences. 
# refer to https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html

# create a figure, in which you have four plots arranged in 2 rows and 2 columnes
# plt.subplots(nrows, ncols)
fig, axs = plt.subplots(2, 2)

# as a result, the returned axs object (a numpy array) has four components
# and each corresponds to a specific location in a row and column
# use index to access a plot (axes) to draw lines, ......
# here, axs is a 2D array. If there's only one row, axs is a 1D array, e.g., fig, ax = plt.subplots(1, 3) => ax[0], ax[1], ...
# axe.plot(), similar to plt.plot(), refer to https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html#matplotlib.axes.Axes.plot
axs[0, 0].plot(x, y, label='line_1')

# create legend for a single plot
# same as plt.legend()
axs[0, 0].legend()

# name x and y for a single plot
# differ from plt.xlabel()
axs[1, 0].set_xlabel('x-axis')
axs[1, 0].set_ylabel('y-axis')

# create title for the entire figure
fig.suptitle('Title')

# create title for a single plot
for i in range(2):
    for j in range(2):
        axs[i, j].set_title(f'axs[{i},{j}]')

plt.tight_layout()
plt.show()
````

### 1.1.5 Other types of plot: scatter, bar, and histogram

````python
# Use multiple plots in one figure as example

fig, axs = plt.subplots(1, 3, figsize=(10, 3))

# scatter plot: displaying a set of points
# refer to https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.scatter.html#matplotlib.axes.Axes.scatter
axs[0].set_title('scatter plot')
x = np.random.rand(20) # create 20 data points
y = np.random.rand(20)
colors = x + y # color as a function of x and y
sizes = 100 * (x + y) # size as a function of x and y
axs[0].scatter(x, y, c=colors, s=sizes)

# bar plot: displaying a sequence of numbers as bars
# refer to https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.bar.html#matplotlib.axes.Axes.bar
axs[1].set_title('bar plot')
height = [20, 4, 15] # height of each bar
x = [1, 2, 3] # position of the bars on x axis 
labels = ['Bar 1', 'Bar 2', 'Bar 3']
axs[1].bar(x, height , tick_label=labels, width=0.6)

# histogram plot: a discretized display of value frequency, a way to visualize distribution
# refer to https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.hist.html#matplotlib.axes.Axes.hist
axs[2].set_title('histogram plot')
data = np.random.randn(1000) # create 1000 random values with Gaussian distribution
axs[2].hist(data, bins=30)

# save figure
# fig.savefig("./output/example.png") # format can also be .jpg, .pdf, ......

plt.show()
````

## 3. Plot points and Multiple Charts

Please run the following cell containing useful functions already implemented for you to plot some charts.

````python
def is_above_fn(p, x, y):
    """ This functions returns True if the point p (as a tuple) is above the line defined by the points in x and y"""
    return np.cross(p-x, y-x) < 0

def find_above_and_below_points_fn(x, y, x_bound, y_bound):
    """ This funtion split a numpy array into two numpy arrays with all the points that stand 
    above and below a line, respetively"""
    X_above = []
    X_below = []
    Y_above = []
    Y_below = []
    for xi, yi in zip(x, y):
        if is_above_fn((xi, yi), x_bound, y_bound):
            X_above.append(xi)
            Y_above.append(yi)
        else:
            X_below.append(xi)
            Y_below.append(yi)
    X_above = np.array(X_above)
    X_below = np.array(X_below)
    Y_above = np.array(Y_above)
    Y_below = np.array(Y_below)
    return X_above, Y_above, X_below, Y_below

def generate_gradient_colors_fn(x, y):
    """ This functin generates color values as a function of x and y"""
    return x + y
````
